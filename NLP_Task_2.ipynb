```python
#importing Then spaCy library
import spacy
nlp = spacy.load("en_core_web_md")
```

```python
# Reading the small english language models

file_path ='/content/basic_nlp_text.txt'
with open(file_path,'r') as file:
  text = file.read()
```

```python
#Processing the text using SpaCy
doc = nlp(text)
print(doc)
```

```python
#Extracting and printing named emtities with their labels
print('Named entities and their Labels:')
for entity in doc.ents:
  print(entity.text,entity.label_)
```
